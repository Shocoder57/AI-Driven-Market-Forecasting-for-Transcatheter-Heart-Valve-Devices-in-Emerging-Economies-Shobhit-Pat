```python
import pandas as pd

raw_df = pd.read_csv("Education_indicators_raw_Data.csv")

raw_df.head()
```

```python
df = raw_df.drop(columns=["Country Code", "Indicator Code"], errors="ignore")

df_long = df.melt(
    id_vars=["Country Name", "Indicator Name"],
    var_name="Year",
    value_name="Value"
)

df_long["Year"] = df_long["Year"].astype(int)

df_long.head()
```

```python
indicator_map = {
    "GDP per capita (current US$)": "GDP_per_Capita",
    "Urban population (% of total population)": "Urban_Population",
    "School enrollment, secondary (% gross)": "Enrollment_Rate",
    "Children out of school, primary": "Out_of_School_Rate"
}

df_long["Indicator"] = df_long["Indicator Name"].map(indicator_map)

df_filtered = df_long.dropna(subset=["Indicator"])

df_pivot = df_filtered.pivot_table(
    index=["Country Name", "Year"],
    columns="Indicator",
    values="Value"
).reset_index()

df_pivot.head()
```

```python
df_pivot.rename(columns={"Country Name": "Country"}, inplace=True)
```

```python
df_pivot["Education_Index"] = (
    df_pivot["Enrollment_Rate"] - df_pivot["Out_of_School_Rate"]
)

df_pivot["Market_Growth_Index"] = (
    0.4 * df_pivot["GDP_per_Capita"] +
    0.3 * df_pivot["Urban_Population"] +
    0.3 * df_pivot["Education_Index"]
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

X = df_pivot[
    ["Enrollment_Rate", "Out_of_School_Rate", "Education_Index",
     "GDP_per_Capita", "Urban_Population"]
]

y = df_pivot["Market_Growth_Index"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

lr = LinearRegression()
lr.fit(X_train, y_train)

rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
```
